### Stock Price Data:

#### I.
First we load & clean data from the folder called 'Car', which contains csv files for each individual automotive company from 2020 to 2021. The data itself is generally clean, but we do need to mange to merge each dataframe into a large dataframe. To achieve that, we create a new variable called 'Maker', which specifies which company the observation belongs to. After the new variable is created for each observation, we concatenate each dataframe to a new dataframe by rows, thus the new dataframe contains all observations from each individual datasets. 

```python
# Read Data from Car folder
data_dir = 'Car'
folder_name = ['BMW','Ford','General Motors','Honda','Nissan','Toyota','Volkswagen']

file_name = []
for name in folder_name:
    if (name == 'Volkswagen'):
        tmp = name
    elif name == 'General Motors':
        tmp = 'gm'
    else:
        tmp = name.lower()
    file_name.append(tmp)

df_name = []

for i,fname in enumerate(file_name):
    file_path = os.path.join(data_dir,folder_name[i])
    file_path = os.path.join(file_path,fname) + '.csv'
    if " " in fname:
        fname = fname.replace(" ","")
    globals()[fname] = pd.read_csv(file_path)
    globals()[fname]['Maker'] = fname
    df_name.append(fname)

df_list = [bmw, ford, gm, honda, nissan, toyota, Volkswagen]
car = pd.DataFrame()

#standardize data sets
for i in df_list:
    i['diff'] = i['Adj Close'] - i['Open']
    for j in ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']:
        i[j] = (i[j]-i[j].mean())/i[j].std()
    
for df in df_list:
    df = df.drop_duplicates(subset=['Date'])
    car = pd.concat([car, df], axis='rows')


car.head()
```

#### II. 
For the data cleaning part, we first filtered out any duplicate observations, which are observations that share the same date, during the process of concatenation. In addition, we filtered out any columns that only have NaN values. Furthermore, given that the dataframe gathers data from 2020 to 2021, there are many observations that dates before the outbreak of Covid-19, thus we filtered out such observations to minimize any outliers that could interfere with our further analysis. Also, we renamed columns of the dataframe to make sure that the naming convention for variables can maintain consistent accross all datasets. Lastly, we sorted the dataframe by date in descending order to make date consistnt accross all datasets.

```python
#clean car dataframe
car = car.dropna(how='all')
car = car[ car['Date'] > '2020-01-16' ]
car = car[ car['Date'] < '2021-03-09' ]
car = car.rename(columns={'Date':'date','Open':'open','High':'high','Low':'low','Close':'close','Adj Close':'adjClose','Volumn':'volumn','Maker':'maker'})
car = car.sort_values(by='date', ascending=False)

car.head()
```

#### The Third outbreak of the epidemic(Sep,2020 to Jan,2021)

During this period, the third outbreak of the epidemic occurred. The number of infections rose from about 50,000 in September to about 70,000 in October, and spiked to around 150,000-200,000 per day for the next four months.
As you can see from our plot of auto share prices, auto share prices differ from our previous understanding that severity of COVID-19 have a strong positive correlation with stock price per share of automotive companies. #### The Ending outbreak of the epidemic(Jan,2021 to Feb,2021)

During this period, the impact of the third outbreak of the epidemic began to gradually end. The number of daily infections has dropped to 50000 from a range of 150,000 - 200,000 in previous months.

In the EDA process, we dissect the time scope(2020-01-17 to 2021-03-07) into four outbreak phases. Given the correlation that we are exploring: the correlation between daily inrease of Covid-19 cases and stock prices of automotive companies, may vary from period to period. We decide to create a new variable that keeps track of each outbreak period.

```python
def outbreak(str):
    if str >= '2020-01-17' and str < '2020-06-31':
        return 'First'
    elif str >= '2020-07-01' and str < '2020-09-32':
        return 'Second'
    elif str >= '2020-10-01' and str < '2021-01-32':
        return 'Third'
    else:
        return 'Fourth'
    
```

```python
merge['outbreak'] = merge['date'].apply(outbreak)
merge.head()
```

With all data merged together and cleaned proporly, we should visualize the data to find any potential relationships. Thanks to our previous EDA analysis, we understand that the distribution of adjusted close price share similar trends accross automotive companies, thus we no longer to differentiate between car brands and explore the generalized correlation between daily increase of Covid-19 and stock prices of automotive companies. To visualize the data, we first used regplot to make 4 subplots, each representing a different outbreak period, to see if there is any potential correlations. 

```python
fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(ncols = 2, nrows = 2, sharey = True, figsize=(20,30))
sns.set(font_scale=1, style="white")

sns.regplot(data = merge[merge['outbreak']=='First'], ax=ax1, x = 'positiveIncrease', y = 'adjClose',line_kws={"color": "red"})
sns.regplot(data = merge[merge['outbreak']=='Second'], ax=ax2, x = 'positiveIncrease', y = 'adjClose',line_kws={"color": "red"})
sns.regplot(data = merge[merge['outbreak']=='Third'], ax=ax3, x = 'positiveIncrease', y = 'adjClose',line_kws={"color": "red"})
sns.regplot(data = merge[merge['outbreak']=='Fourth'], ax=ax4, x = 'positiveIncrease', y = 'adjClose',line_kws={"color": "red"})

plt.suptitle('Correlation between Daily Increase with Stock of Each Outbreak Period', fontsize=20)
fig.subplots_adjust(top=0.95)
ax1.title.set_text('First Outbreak')
ax2.title.set_text('Second Outbreak')
ax3.title.set_text('Third Outbreak')
ax4.title.set_text('Fourth Outbreak')
```

Although there seems to have a correlation between daily increase of Covid-19 and adjusted close stock price of car companies during the third outbreak period, the rest of plots show little to no direct correlation, especially in the fourth period where dotts are rather sparse and seem to have no obvious distribution. To further explore any potential correlations, we use a scatter matrix, which contains daily increase of Covid-19 cases as well as adjusted close prices of stocks of automotive companies. 

```python
sns.pairplot(merge, hue = 'outbreak', vars = ["adjClose", "positiveIncrease"], palette = 'Paired', height = 4)
```

Although the means for adjusted close price are quite similar for the first 3 outbreaks, that of the 4th outbreak differs drastically. In addition, daily increases also differ quite differently in differet outbreak period. And the distribution for adjusted close price and daily increase of Covid-19 do not align. Thus, there doesn't appear to be a strong correlation between stock price of automotive companies with daily increase of Covid-19. Given the scatter matrix, it seems that daily increase of Covid-19 cases does not affect the stock price of automotive companies. 

In order to further analyze whther or not a significant relationship exists between daily increase of Covid-19 with stock prices of car companies, we'll perform linear regression on the adjusted close price and daily increase of Covid-19 for each outbreak period.

```python
for i in ['First', 'Second', 'Third', 'Fourth']:
    print('Outbreak Period:', i)
    reg_df = merge[['adjClose', 'positiveIncrease', 'outbreak']]
    reg_df = reg_df[ reg_df['outbreak']==i ]
    outcome, predictors = patsy.dmatrices("positiveIncrease ~ adjClose", reg_df)

    mod_log = sm.OLS(outcome, predictors)

    fit_res = mod_log.fit()

    print(fit_res.summary())
    print("\n")
```

Based off the p-value returned by linear regression analysis for each outbreak period, we can see that the p-value for the first, second and third outbreak period is relatively small, thus we can consider the increase of daily cases of Covid-19 to be a statistically significant predictor of stock prices of automotive companies. However, in the fourth outbreak period, the p-value is rather large and fails to indicate the existence of correlation. 

Overall, we can see that even though the correlation between the increase of daily cases of Covid-19 and stock prices of automotive companies may not be consistent throughout the time scope of our study, the increase of daily cases of Covid-19 can nonetheless be seen as a statistically significant predictor of stock prices of automotive companies during the period Jan 2020-Jan 2021. However, the p-value in regression analysis that represents the correlation in the fourth outbreak period indicate that seeing the result as it is is quite likely(30.4%) due a random chance alone, which does not support the hypothesis that daily increase of Covid-19 cases has a correlation with stock price of automotive companies. Nonetheless, as for outbreak period 1, 2 and 3, the regression analysis combined with scatter matrix does indicate a strong correlation between daily increase of Covid-19 cases and stock price of automotive companies during the period from Jan 2020 to Jan 2021.
